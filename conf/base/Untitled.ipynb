```python
def preprocess_titanic(df: pd.DataFrame):
    """ This function cleanses the Titanic raw_data, removing unused columns. """
    
    # Remove unused columns 
    df = df.drop(columns=['Name','Ticket','PassengerId','Cabin'])
    
    return df
```

```python
def convert_sex_number(df: pd.DataFrame):
    """ This function converts the sex of the person to a integer 
        in two separated columns. """
    
    df_1 = pd.get_dummies(df, columns=['Sex'])
    df = df_1.rename(columns = {'Sex_0':'Male', 'Sex_1':'Female'})
    
    return(df)
```

```python
def fixing_embarked(df: pd.DataFrame):
    """ This function converts the null embark data to the most 
    commom port of embarkation. """
    
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
    
    return df
```

```python
def convert_embarked_number(df: pd.DataFrame):
    """ This function converts the embarked data to a integer 
        in three separated columns. """
    
    df = pd.get_dummies(df, columns=['Embarked'])
    df = df.rename(columns = {'Embarked_C':'Emb_C', 'Embarked_Q':'Emb_Q', 
                              'Embarked_S':'Emb_S'})
    
    return df
```

```python
def cabin_string (df:pd.DataFrame):
    """ This function removes the numbers in the cabin string and changes
    the column to 'Section'. """
    
    df['Section'] = df['Cabin'].apply(lambda x: x[0])
    df['Section'] = df['Section'].replace('n', np.nan)
    df = df.drop(columns=['Cabin'])
    
    return df
```

```python
def convert_section_number(df: pd.DataFrame):
    """ This function converts the section data to a integer 
        in separated columns. """
    
    df = pd.get_dummies(df, columns=['Section'], dummy_na= True)
    df = df.rename(columns = {'Section_A':'Sec_A', 'Section_B':'Sec_B', 
                              'Section_C':'Sec_C', 'Section_D':'Sec_D',
                              'Section_E':'Sec_E', 'Section_F':'Sec_F',
                              'Section_G':'Sec_G', 'Section_T':'Sec_T',
                              'Section_nan':'Sec_nan'
                             })
    
    return df
```
